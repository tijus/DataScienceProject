```r
# Problem - 6
library("ggmap")
library("maptools")
library(maps)
visited <- c("SFO", "Chennai", "London", "Melbourne", "Lima,Peru", "Johannesbury, SA")
ll.visited <- geocode(visited)
visit.x <- ll.visited$lon
visit.y <- ll.visited$lat
map("world", fill=TRUE, col="white", bg="lightblue", ylim=c(-60, 90), mar=c(0,0,0,0))
points(visit.x,visit.y, col="red", pch=36)

visited1 <- c("SFO", "New York", "Buffalo", "Dallas, TX")
ll.visited1 <- geocode(visited1)
visit1.x <- ll.visited1$lon
visit1.y <- ll.visited1$lat
map("state", fill=TRUE, col=rainbow(50), bg="lightblue", mar=c(0,0,0,0))
points(visit1.x,visit1.y, col="yellow", pch=36)
```

```r
# Problem - 8
# Reference -> R for DataScience (http://r4ds.had.co.nz/graphics-for-communication.html)
# Installing ggplot2, ggrepel, viridis
# install.packages("ggplot2", "ggrepel", "viridis")
# ggplot2 is a very good tool for visualizing the results of exploratory data analysis
# The below example shows one such visualization
library(ggplot2)
data(mpg)
summary(mpg)
# The ggplot takes in the mpg dataset form ggplot2 library and create a graph according to the two axis fuel efficiency(displ), and engine size(hwy)
# This deviation is shown based on the third attribute, class of automotives
 ggplot(mpg, aes(displ, hwy)) + 
    geom_point(aes(color = class)) + 
    geom_smooth(se = FALSE) + 
    labs(title = "Fuel efficiency generally decreases with engine size")

#In this section, we are using ggrepel, ggplot2 extension library which is for separating the closely packed labels
# The usage of this library is shown below
# Using ggplot extension package ggrepel or geom_text_repel
library(ggrepel) # Loading the library
set.seed(42) # Setting the seed point
# plotting mpg mtcars dataset using as a point and using ggrepel to avoid overlapping of labels
ggplot(mtcars) +
  geom_point(aes(wt, mpg), color = 'red') +
  geom_text_repel(aes(wt, mpg, label = rownames(mtcars))) +
  theme_classic(base_size = 16)

# using another ggplot2 extension library qqplotr
# Reference: qqplotr vignette <- https://github.com/aloy/qqplotr/blob/master/vignettes/introduction.Rmd

require(qqplotr) # loading qqplotr library
set.seed(0) # setting the seed point to 0
smp <- data.frame(norm = rnorm(100)) # data simulation from standard normal distribution
gg <- ggplot(data = smp, mapping = aes(sample = norm)) + # Initializing the normal distribution 
    geom_qq_band(bandType = "ts", mapping = aes(fill = "TS")) + # Initializing quartile-quartile plot
    geom_qq_band(bandType = "normal", mapping = aes(fill = "Normal")) +
    geom_qq_band(bandType = "boot", mapping = aes(fill = "Bootstrap")) +
    stat_qq_line() + # initializing stat line
    stat_qq_point() + # initializing stat point
    labs(x = "Theoretical Quantiles", y = "Sample Quantiles") + # providing labels
    scale_fill_discrete("Bandtype") 
gg



```
