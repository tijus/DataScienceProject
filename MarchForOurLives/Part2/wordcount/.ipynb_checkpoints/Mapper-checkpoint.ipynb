```python
import sys
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import pandas as pd 
stop_words = set(stopwords.words('english'))
```

```python
def reg (line):
    line = (re.sub('(RT )*@.+? ', '',line)) # dont want <RT and/or @user> tags
    line = (re.sub('http.+? ', '',line)) # dont want web adresses
    line = re.sub(',', ' ', line)
    line = line.lower()
    words = re.findall('([a-z]+?.{0,1}?[a-z]*?)\W*\s+',line) # regex to extract words (space delmit, remove special chars at end of words)
    return words
#reg(Tooter)
```

```python
def mapper(tweet):
    
    lines = tweet.split('\n')
    print(lines)

    for line in lines:
    #print(line)
    # here I am going to do pre processing using regex 
        words = reg(line)
    #print(words)

    # filtered_sentence = [w for w in words if not w in stop_words]
        filtered_sentence = []

    for w in words:
        if w not in stop_words:
            filtered_sentence.append(w)

    #for word in filtered_sentence:
     #   print(word+"\t"+str(1))

    return
```
